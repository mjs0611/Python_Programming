```
"""
조건문: 조건에 맞는 경우에만 실행하는 문장
if 조건식:
  조건이 참(True)일 때 실행할 문장들
"""

age = int(input("나이 입력 >>> "))
if age >= 0:   # age의 값이 0보다 크거나 같으면
  # 조건식이 참일 때 실행하는 블록(block)
  print("나이:", age)
print("코드 끝")
```

```
"""
if 조건식:
  조건식이 참일 때 실행할 문장들
else:
  조건식이 거짓일 때 실행할 문장들
"""

# 정수 입력 -> 짝수이면 "even"을 출력, 그렇지 않으면 "odd"를 출력
number = int(input("정수 입력 >>> "))
if number % 2 ==0:
  print("even")
else:
  print("odd")
```

```
"""
if 조건식1:
  조건식1이 참일 때 실행할 문장들
elif 조건식2:
  조건식2가 참일 때 실행할 문장들
...
else:
  위의 모든 조건이 거짓일 때 실행할 문장들
"""

number = int(input("정수 입력 >>> "))
if number > 0:
  print("positive")
elif number == 0:
  print("zero")
else:
  print("negative")
```

```
# 시험 성적(정수) 입력
# 90점 이상이면 -> A
# 80점 이상이면 -> B
# 70점 이상이면 -> C
# 70점 미만이면 -> F

score = int(input("시험 점수 입력 >>> "))
if score >= 90:
  print("A")
elif score >= 80:
  print("B")
elif score >= 70:
  print("C")
else:
  print("F")
```

```
# 나이 입력
# 입력받은 나이가 0세 이상 200세 미만이면, "정상입력" 출력
# 그렇지 않으면, "잘못 입력했습니다." 출력

age = int(input("나이 입력 >>> "))
if 0 <= age < 200:   # (age >= 0) and (age < 200):
  print("정상입력")
else:
  print("잘못 입력했습니다.")
```
