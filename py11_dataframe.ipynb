```
import pandas as pd
```

```
# CSV(Comma-Separated Values) : 
# 값들이 comma(,)로 구분되어 있는 텍스트 파일
# 구분자(separator) : 값들을 구별하기 위해서 넣는 특수 문자
# CSV 파일 일반적으로 comma를 구분자로 사용하지만,
# 탭('\t'), 콜론(:), 이중콜론(::) 등 다양하게 구분자를 사용할 수 있음

path = 'drive/My Drive/Colab Notebooks/202006/gapminder.tsv'
gapminder = pd.read_csv(path, sep='\t')
gapminder.head()   # DataFrame에서 첫 5개 행만 선택
```

```
print(gapminder.shape)   # (row 개수, column 개수)
```

```
print(gapminder.columns)   # 컬럼 이름
```

```
gapminder.describe()   # 요약 기술 통계량(descriptive statistics)
```

```
# DataFrame에서 컬럼을 선택 : df[컬럼이름], df[[col1, col2, ...]]
# gapminder DF에서 나라이름, GDP 선택
gapminder[["country", "gdpPercap"]]
```

```
# DataFrame에서 row로 선택 : df.loc[], df.iloc[]
gapminder.loc[0:4]
```

```
gapminder.iloc[0:5]
```

```
# gapminder DF에서 끝에서 7개의 행(row)를 선택
gapminder.iloc[-7:]
```

```
# boolean indexing
# gapminder 데이터프레임에서 2007년 자료만 선택
gapminder["year"] == 2007
```

```
gapminder[gapminder["year"] == 2007]
```

```
# Asia 대륙의 자료만 선택
gapminder[gapminder["continent"] == "Asia"]
```

```
# Asia 대륙의 2007년 자료만 선택
# () & (), () | (), ~()
gapminder[(gapminder["continent"] == "Asia") & (gapminder["year"] == 2007)]
```

```
# Asia 대륙에서 1인당 GDP(gdpPercap)가 10,000 이상인 자료들만 선택
gapminder[(gapminder["continent"] == "Asia") & (gapminder["gdpPercap"] >= 10_000)]
```

```
# 대한민국의 1인당 GDP가 10,000 이상인 자료들
gapminder[(gapminder["country"] == "Korea, Rep.") & (gapminder["gdpPercap"] >= 10_000)]
```

```
gapminder[(gapminder["country"].str.contains("Korea")) & (gapminder["gdpPercap"] >= 10_000)]
```

```
# Asia 대륙에서 1인당 GDP가 20,000 이상인 자료에서
# 나라 이름, 연도, 1인당 GDP를 출력
subset = gapminder[(gapminder["continent"] == "Asia") & (gapminder["gdpPercap"] >= 20_000)]
subset
```

```
subset[["country", "year", "gdpPercap"]]
```
