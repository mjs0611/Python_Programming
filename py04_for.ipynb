```
# 문자열은 +, * 연산이 가능
# +: 두개의 문자열을 이어주는 기능
# *: 문자열을 반복해서 늘려주는 기능
print("abc" + "def")
print("abc" * 3)
```

```
# range(start, end, step)
# start <= x < end 범위 안의 숫자들을 만들어주는 기능
# step만큼씩 자동으로 늘려줌. step이 1일 때는 생략 가능
for x in range(1, 6):
  print(x)
```

```
for x in range(5, 0, -1):
  print(x)
```

```
# range(end): 0 <= x < end 미만의 1씩 증가하는 숫자들
for x in range(3):
  print(x)
```

```
# 빈 리스트(empty list) 생성
numbers = []
print(numbers)
numbers.append(1)
print(numbers)
numbers.append(2)
print(numbers)
```

```
# 빈 리스트에 10 이하의 양의 정수들을 저장 : [1, 2, 3, ..., 10]
numbers = []
for n in range(1, 11):
  numbers.append(n)

print(numbers)
```

```
# 빈 리스트에서 10 이하의 양의 정수들 중에서 짝수만 저장 : [2, 4, 6, 8, 10]
numbers = []
for n in range(2, 11, 2):
  numbers.append(n)

print(numbers)
```

```
numbers = []
for n in range(1, 11):   # 1 <= n < 11 범위의 정수들
  if n % 2 == 0:   # 2로 나눈 나머지가 0이면(짝수이면)
    numbers.append(n)
  print(numbers)
```

```
# break : 반복문을 중간에 종료하고 싶을 때 사용
for n in range(10):   # 0 <= n < 10 범위의 정수들 중에서
  print(n)
  if n >= 5:   # 만약 n이 5이상이 되면
    break   # 반복문을 멈춤
```

```
# continue : continue 아래의 문장들을 실행하지 않고,
# 다음 반복을 위해서 반목문의 시작부분으로 되돌아 가는 기능
for n in range(5):   # 0 <= n < 5 범위의 정수들 중에서
  if n == 3:
    continue
  print(n)
```

```
# dict : key-value 쌍으로 이루어진 데이터들을 저장하는 자료 타입
cities = {"Seoul" : "Korea",
          "Washington" : "USA",
          "London" : "UK"}
for capital in cities:
  print(capital, cities[capital])
# for-in 구문에서 dict를 사용하면, dict의 키를 차례대로 추출하는 동작을 반복
```

```
# dict의 value들만 선택하고 싶을 때
for country in cities.values():
  print(country)
```

```
# dict에서 key, value 모두 반복하고 싶을 때:
for city, country in cities.items():
  print(city, country)
```

```
# list comprehension
numbers = [1, 2, 3, 4, 5, 6]
# numbers 안에 숫자들 중에서 짝수들로만 이루어진 새로운 리스트 evens 생성
# evens = []
# for x in numbers:   # numbers 안에 숫자들 중에서
#   if x % 2 == 0:   # x가 짝수이면
#     evens.append(x)
evens = [x for x in numbers if x % 2 == 0]


print(evens)

# Ctrl + / : 주석 토글 기능
```

```
genders = ["male", "female", "female", "male"]
gender_code = []   # male - 0, female - 1
for g in genders:
  if g == "male":
    gender_code.append(0)
  else:
    gender_code.append(1)
print(gender_code)
```

```
gender_code2 = [0 if gender == "male" else 1
                for gender in genders]
print(gender_code2)
```

```
# 2중 반복문
for x in range(1, 10):
  for y in range(1, 10):
    print(f"{x} X {y} = {x * y}")
  print("-" * 20)
```

```
# dictionary coprehension
data = {}   # empty dict
print(data)
data[1] = "apple"   # dict_name[key] = value
print(data)
data[100] = "MS"
print(data)
```

```
product_id = [100, 101, 200, 201]
product_name = ["iphone", "galaxy", "TV", "Blueray"]
for id_, name in zip(product_id, product_name):
  print(id_, name)
```

```
products = {}   # empty dict
for id_, name in zip(product_id, product_name):
  products[id_] = name   # 빈 dict에 새로운 원소(key-value) 추가
print(products)
```

```
# dict comprehension
products2 = {key : value for key, value in zip(product_id, product_name)}
print(products2)
```
